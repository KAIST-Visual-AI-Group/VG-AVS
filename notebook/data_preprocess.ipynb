```python
import json

json_path = "/home/daehyeonchoi/embodied/AVS_release/data/avs_procthor_train.jsonl"

src_root = "/home/daehyeonchoi/embodied/procthor-10k/outputs/procthor_active_vqa_existence_1106"
tgt_root = "avs_procthor_train"

dataset = []
with open(json_path, "r") as f:
    for line in f:
        data = json.loads(line)

        for k, v in data.items():
            if isinstance(v, str) and v.startswith(src_root):
                data[k] = v.replace(src_root, tgt_root)

        dataset.append(data)

with open(json_path, "w") as f:
    for data in dataset:
        f.write(json.dumps(data) + "\n")

```
